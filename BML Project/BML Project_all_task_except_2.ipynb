```python
# Cost Function

def computeCost(X,y,theta):
    
    """This function optimizes the weights
    for every feature in the training dataset.
    The optimization technique used in this 
    case is Mean Squared Error (MSE)
    
    Parameters
    -----------
    Feature matrix 'X', Target values 'y', 
    Weights and Bias matrix 'theta'
    
    Returns
    -------
    Mean Squared Error"""
    
    tobesummed = np.power(((X @ theta.T)-y),2)     # Calculates difference between target value 'y' and calculated value 'X @ theta.T'
    return np.sum(tobesummed)/(2 * len(X))         # Summation of differences divided by 2 * length of X
```

```python
import timeit
start_time = timeit.default_timer()                 # Runtime counter starts

# Gradient Descent Function

def gradientDescent(X,y,theta,iters,alpha):
    
    """This function minimizes the MSE value
    by calculating the gradient of cost function. 
    For this, we calculate the partial derrivatives
    in terms of weight and bias.
    
    Parameters
    ----------
    Feature matrix 'X', Target values 'y', 
    Weights and Bias matrix 'theta', 
    hyperparameters 'iters' and 'alpha'
               
    Returns
    -------
    updated 'theta' and corresponding 'cost' value. """
    
    cost = np.zeros(iters)
    for i in range(iters):
        theta = theta - (alpha/len(X)) * np.sum(X * (X @ theta.T - y), axis=0)
        cost[i] = computeCost(X, y, theta)
    
    return theta,cost

# Training model using feature dataset 

g,cost = gradientDescent(X_train,y_train,theta,iters,alpha)
print(g)

finalCost = computeCost(X_train,y_train,g)
print(finalCost)

elapsed = timeit.default_timer() - start_time      # Runtime counter stops
print('Runtime: ' , elapsed, ' seconds')
```

```python
def e_func(x):
    
    s_x1 = np.sqrt(1.0625)
    s_x2 = np.sqrt(1.0625)
    ro = 1/1.0625
    mu_x1 = 1
    mu_x2 = 1
    p0 = np.log(2*np.pi*s_x1*s_x2*np.sqrt(1-ro**2))
    p1 = 1/(2*(1-ro**2)) * (((x[0]-mu_x1)**2/s_x1**2) + 
                            ((x[1]-mu_x2)**2/s_x2**2) - 
                            ((2*ro*(x[0]-mu_x1)*(x[1]-mu_x2))/(s_x1*s_x2)))
    lgp = p0 + p1
    return lgp
```

```python
def e_grad(x):
    g = np.empty(2)
    s_x1 = np.sqrt(1.0625)
    s_x2 = np.sqrt(1.0625)
    ro = 1/1.0625
    mu_x1 = 1
    mu_x2 = 1

    g[0] = (2*(x[0]-1) - 2*ro*(x[1]-1)) / (1.0625*2*(1-ro**2))
    g[1] = (2*(x[1]-1) - 2*ro*(x[0]-1)) / (1.0625*2*(1-ro**2))
    return g
```

```python
def e_func(x, f):
    # x[0]->x[7] weights
    # x[8] bias
    # x[9] log alpha
    # x[10] log sigma^2
    
    x_train = f[0]
    t_train = f[1]
    
    N = len(t_train)
    p0 = N/2*(np.log(2*np.pi) + x[10])
    
    p1 = 0
    for n in range(N):
        summ = 0
        for m in range(0,9):
            summ += x[m]*x_train[n][m]
        
        diff = t_train[n] - summ
        p1 += diff
            
    p2 = 0
    for m in range(0,9):
        diff2 = 0.5*(x[9]-np.log(2*np.pi)) - np.exp(x[9]*x[m]*0.5)
        p2 += diff2
    
    lgp = -p0 - p1 +p2
    return -lgp
```

```python
def e_grad(x, f):
    # x[0]->x[7] weights
    # x[8] bias
    # x[9] log alpha
    # x[10] log sigma^2
    
    x_train = f[0]
    t_train = f[1]
    weights = np.array([x[i] for i in range(0,9)])
    s2 = np.exp(x[10])
    alpha = np.exp(x[9])
    
    N = len(t_train)
    g = np.empty(11)
    
    # weights and bias gradients
#     for i in range(0,9):
#         g[i] = 0
#         for n in range(N):
#             summ = 0
#             for m in range(0,9):
#                 summ += x[m]*x_train[n][m]
        
#             diff = t_train[n] - summ
#             g[i] += diff
            
#         g[i] /= 2*np.exp(x[10])
#         for m in range(0,9):
#             mul = -np.exp(x[9])*x[m]
            
#         g[i] += mul  

#     print(1/s2)
#     print(sum((t_train - weights@x_train.T) @ x_train))
#     print(sum(alpha*weights))
    g[:9] = 1/s2 * sum((t_train - weights@x_train.T) @ x_train) + sum(alpha*weights)
        
    
    # log alpha gradient 
    g[9] = 0
    for m in range(0,9):
        g[9] += 0.5*(1-np.exp(x[9])*(x[m]**2)) 
        
    # log sigma^2 gradient
    g[10] = -N*0.5
    for n in range(N):
        summ = 0
        for m in range(0,9):
            summ += x[m]*x_train[n][m]
        
        diff = t_train[n] - summ
        g[10] += diff
            
    g[10] /= 2*np.exp(x[10])  

    
    return g
```

```python
def e_func(x, f):
    # x[0]->x[7] weights
    # x[8] bias
    # x[9] log alpha
    # x[10] log sigma^2
    # x[11]->x[18] alpha_m for weights
    # x[19] alpha_m for bias
    
    x_train = f[0]
    t_train = f[1]
    weights = np.array([x[i] for i in range(0,9)])
    alpha_m = np.array([x[i] for i in range(11,20)])
    s2 = np.exp(x[10])
    
    N = len(t_train)
    p0 = N/2*(np.log(2*np.pi) + x[10])
    
    p1 = 0
    for n in range(N):
        summ = 0
        for m in range(0,9):
            summ += x[m]*x_train[n][m]
        
        diff = t_train[n] - summ
        p1 += diff
            
    p2 = 0
    for m in range(0,9):
        diff2 = 0.5*(alpha_m[m]-np.log(2*np.pi)) - np.exp(alpha_m[m]*weights[m]*0.5)
        p2 += diff2
    
    lgp = -p0 - p1 +p2
    return -lgp
```

```python
def e_grad(x, f):
    # x[0]->x[7] weights
    # x[8] bias
    # x[9] log alpha
    # x[10] log sigma^2
    # x[11]->x[18] alpha_m for weights
    # x[19] alpha_m for bias
    
    x_train = f[0]
    t_train = f[1]
    weights = np.array([x[i] for i in range(0,9)])
    alpha_m = np.array([x[i] for i in range(11,20)])
    s2 = np.exp(x[10])
    
    N = len(t_train)
    g = np.empty(20)
    
    # weights and bias gradients

    g[:9] = 1/s2 * sum((t_train - weights@x_train.T) @ x_train) + sum(alpha_m*weights)
        
    
    # log alpha gradient 
    g[9] = 0
    for m in range(0,9):
        g[9] += 0.5*(1-np.exp(x[9])*(x[m]**2)) 
        
    # log sigma^2 gradient
    g[10] = -N*0.5
    for n in range(N):
        summ = 0
        for m in range(0,9):
            summ += x[m]*x_train[n][m]
        
        diff = t_train[n] - summ
        g[10] += diff
            
    g[10] /= 2*np.exp(x[10]) 
    g[11] = 0
    
    # alpha_m gradient
    for m in range(1,9):
        g[(m+11)] = np.abs(0.5*(1/np.exp(alpha_m[m]) - weights[m]**2/np.exp(alpha_m[m])))
    
    return g
```
